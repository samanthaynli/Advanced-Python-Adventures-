### copy over data cleaning and normalizing from W6

```python
df2=df.drop(columns=['embarked'])
df2=df2.replace('?',np.nan) #replace the ? string with not a number 
df2['gender']=df2['gender'].replace({'female':0,'male':1}) #convert gender to digits
df2['gender']=pd.to_numeric(df2['gender'],errors='coerce') #replaces unknown values with NaN not throw an error
df2['gender']=df2['gender'].fillna(df2['gender'].mode()[0]) #supposedly replaces the NaN in row 4 with the most common gender but I literally can't figure it out
df2['age']=pd.to_numeric(df2['age'],errors='coerce') #replaces unknown values with NaN not throw an error
df2['age']=df2['age'].fillna(df2['age'].median()) #convert all NaN to median (convention)
df2['fare']=pd.to_numeric(df2['fare'],errors='coerce') #replaces unknown values with NaN not throw an error
df2['fare']=df2['fare'].fillna(df2['fare'].median())
df2['family']=df2['family'].fillna(df2['family'].median())
df2.head(10)
```

### Split data

```python
X=df2.drop(columns=['Unnamed: 3','date','survived'])
y=df2['survived']
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2,random_state=67)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
log_reg = LogisticRegression()

param_grid = {
'penalty': ['l1', 'l2'],
'C': [0.001, 0.01, 0.1, 1, 10, 100],
'solver': ['liblinear'] # 'liblinear' works well for small/medium datas
}
grid_log = GridSearchCV(log_reg,param_grid,cv=5,scoring='accuracy')
grid_log.fit(Xtrain, ytrain)
print(f"Best Penalty: {grid_log.best_params_['penalty']}")
print(f"Best C: {grid_log.best_params_['C']}")
best_log_model = grid_log.best_estimator_
```
