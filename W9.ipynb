```python
from sklearn.datasets import load_sample_image
import matplotlib.pyplot as plt
china = load_sample_image("china.jpg")
plt.imshow(china)
plt.axis('off')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_sample_image
pic = load_sample_image("china.jpg")
A = pic.reshape(-1, 3)
X2 = A.copy()
km2 = KMeans(16, random_state=67)
km2.fit(X2)
sci_image = km2.cluster_centers_[km2.labels_].reshape(pic.shape)
sci_image = sci_image.astype(np.uint8)
plt.imshow(sci_image)
plt.axis('off')
plt.show()
```
